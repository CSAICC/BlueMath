```python
import xarray as xr
import pandas as pd
```

```python
subset_mda = xr.open_dataset("inputs/subset_MDA.nc")
subset_mda["freq"] = 1.0 / subset_mda["Tp"]
subset_mda
```

```python
eofs = pd.read_pickle("inputs/eofs.pkl")
eofs
```

```python
dems = xr.open_dataset("inputs/dems_laredo_D3D_processed.nc")
mean_bat = dems["elevation"].mean(dim="toba")
mean_bat
```

```python
fixed_parameters = {
    "subset_mda": subset_mda,
    "eofs": eofs,
    "dems": dems,
    "mean_bat": mean_bat,
}
metamodel_parameters = {
    "case_index": list(range(1000)),
    "swl": subset_mda.SWL.values,
}
```

```python
from bluemath_tk.wrappers.xbeach.xbeach_example import ManuXBeachModelWrapper


xbeach_model = ManuXBeachModelWrapper(
    templates_dir="templates",
    metamodel_parameters=metamodel_parameters,
    fixed_parameters=fixed_parameters,
    output_dir="outputs",
)
```

```python
xbeach_model.build_cases()
```
